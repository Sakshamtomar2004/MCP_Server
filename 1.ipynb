```python
from mcp.server.fastmcp import FastMCP
```

```python
mcp=FastMCP("Math")
@mcp.tool()
def add(a:int,b:int)->int:
    """Adding two numbers"""
    return a+b


@mcp.tool()
def multiply(a:int,b:int)->int:
    """Multiply two numbers"""
    return a*b
```

```python
if __name__=="__main__":
    mcp.run(transport="stdio")
```
